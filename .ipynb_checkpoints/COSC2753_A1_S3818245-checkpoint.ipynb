## II) Approach

### Choosing Machine Learning models
Up until now, there are 5 types of supervised machine learning models that we have covered and will be evaluated for this problem:  

- **Linear Regression** - Hypothesis function is a first-degree function. Used for predicting *continuous* variable. The output from a Linear Regression model is the predicted value of dependent variable (label) based on the given values of features (unseen input).  

- **Polynomial Regression** - Hypothesis function is a n-th degree function. Similar to Linear Regression, it is also used for predicting *continuous* variable. The output from a Polynomial Regression model is the predicted value of dependent variable (label) based on the given values of features (unseen input).  

- **Logistic Regression** - Hypothesis function is a sigmoid function which output value ranging from 0-1. This model is used for predicting *discrete* variable (classification problem). The output from a Logistic Regression model gives out binary output (either 0 or 1).

- **Decision Tree Classifier** - Use a tree structure of multiple nodes where each node is a decision point to split data to different branches. Stop splitting (reach leaves) when data is classified to required classes.

- **Random Forest** - Combine multiple trees for classification, each tree is constructed with a different sub-set of training data and features. Final classification is decided by majority voting.


Since we are only trying to predict *whether a patient get Sepsis or not*, it falls into a classification problem (discrete output), thus **Logistic Regression** and **Decision Tree Classifier**/**Random Forest** are the model that will be considered for this problem. 

From the **EDA** section, we have concluded that none of the features have a clear decision boundary (range of value that can indicates patient is positve to Sepsis or not) from **scatter plot** and **box plot**. Even with the **pair plot** which consider a pair of value, no clear separation can be seen. 

Since most of features/ pair of features does not have clear separation, it can be predicted that the performance of **Logistic Regression** and **Decision Tree Classifier**/**Random Forest** would be close to each other. Thus, the final model is the better performing model between **Logistic Regression** and **Random Forest**.

### Choosing Input Features   

Another issue that we have to consider is choosing between 2 approaches of analysis:

- **Univariate** - 1 input attribute (feature)
- **Multivariate** - multiple input attributes (features)

From the EDA process in the previous section, we can conclude that **PL** have the strongest correlation with the result for **Sepsis** thus it is guaranteed to be one of the features for **Logistics Regression**. 

Due the nature of a decision tree where accurate classification requires multiple features (for multiple decision nodes) to reduce impurity, **Univariate** would not be tested for **Random Forest**.

Other than that, a test with all features would be also carried out to check which features affect the result/classification the most (by looking at **coefficient** for **Logistics Regression** and **feature importance** in **Random Forest**). 

In conclusion, to ensure that the model being used is the best possible, we would evaluate the output of **2 different input**:
- A **Univariate** model with **PL** being the only input attribute for **Logistic Regression**.
- A **Multivariate** model with all 11 features as input attributes for **Logistic Regression** and **Random Forest**.

### Model evalutation methods

- For **Logistic Regression**, a **loss function** (**Log loss**) is used to see how accurate the model prediction is compared to the expected output. 
- As this is a classification problem, metrics such as **Recall**, **Precision**, **F1**, **ROC AUC** can also be used to measure the performance of models in various aspects (how well it classify each class).
- **Confusion matrix** can be used to show how the data is classified.
- As stated in the Problem Statement, our aim is to *save resources* and *prevent the risk of health complication*. Between the 2, it is clear that the priority is *prevent the risk of health complication* meaning the number of ***False Negative*** (Positive but classified as Negative) must be kept to the lowest, meaning the metrics that would be prioritized/ maximized are **Recall** for **Positive** and **Precision** for **Negative**. To make sure that the other requirement (*save resources*) is maintained, **F1** is also an important metrics to ensure the balance.
- **K-fold cross-validation** (use different part of data for validation each fold) is also used for validation process.

---

### Regularization (for Logistic Regression)

As mentioned in a discussion regarding the tendency to overfit of **Logistic Regression** on Stack Exchange \[1\], the model would usually try to maximize weights to get ouput closer to 0 and 1 as possible. Thus, **Regularization** is required to prevent overfitting/overly large weights. 

There are **3 regularization methods** for Logistic Regression mentioned in the 3rd lecture \[2\] and Google's Logistics Regression lesson \[3\]:

> - L1 (Lasso Regression) - some weights maybe zero, meaning it would also conduct feature selection. Penalize by adding sum of absolute value of weights to the loss function. It is also robust against outliers. (definition from Kurtis Pykes \[4\])   
> - L2 (Ridge Regression) - all weigths will be non-zero but some will be very small. Penalize by adding sum of squared value of weights to the loss function (higher penalty for outliers). (definition from Kurtis Pykes \[4\])   
> - Early stopping, meaning to reduce training steps. 

First of all, for early stopping regularization approach, there is no clear indication to where a model should stop training and it should be based on experience. Since it is a risky approach and it can greatly affect the model output if the training is not enough, this method would not be used in this project.  

From 'Choosing Input' section, we have established that we will try out 2 different input features. As such, regularization method would also varies as follow:
- **Univariate** with **PL** : as there is no need for feature selection, and we also want to avoid the model set the only feature to 0 (exclude the only feature), we would apply **L2** for this model.
- **Multivariate** with all features: In this case, since our dataset have the traits that suitable to both regularization methods (features with many outliers such as **TS**, **BD2** might do well with **L1**, however, since the dataset have few features, feature selection from **L1** might affect the result), hence both **L1** and **L2** are tested and compared.

---

### Pruning (for Decision Tree Classifier/ Random Forest)
- **Post-pruning** - the method provided by sklearn is **Minimal Cost-Complexity Pruning**, to choose the best level of pruning, parameter tuning (**ccp_alpha**) is applied (sklearn documentation \[5\]).
- **Pre-pruning** - similarly, there are several tunable parameter is provided by sklearn for Pre-pruning which are **min_impurity_decrease**, **min_samples_split**, **min_samples_leaf** that determine the minimal impurity needed for a node to be split (suggested by Sharoon Saxena \[6\]).

## III) Evaluating model

### 1) Prepare for training

### Training data preprocessing

One of the conclusions from the EDA is that the range of data is starkly different in various features (especially **TS** value ranging from **0 - 846** and **M11** ranging from **0 - 2.42**). Another issue of the data is that there is positive skewness detected in various feature. 

Thus as mentioned in a post on Kaggle comparing **Scaling** and **Normalization** \[7\] the training data would be **normalized** before training process since it also address/adjust the skewness of the dataset as opposed to scaling which only adjust the range of data.  

As explained on a discussion on Stack Overflow \[8\], **row normalization** is useful only when the features don't have unit (so data on different columns can be adjusted to the same scale without affecting to much) and there is a risk of information loss. Since the unit in the dataset is useful, **column normalization** is applied.

As mention by user Hong Ooi on Stack Overflow \[9\], **normalization** is ***only*** applied for data trained with Logistics Regression. Due to the nature of Random Forest that only classify based on threshold thus normalize would not contribute to the result and instead require test data to also be normalized.

Another issue about the dataset addressed in **EDA** is that there is an imbalance between **Positive** and **Negative** classes, **Oversampling technique** is applied. Particularly, the algorithmn being used is **SMOTE** (Synthetic Minority Oversampling Technique) which generate data points that connects from inliers to outliers to ensure that the distribution (range of value) of the data is still in-tact. This method is chosen as opposed to Random Oversampling which gives random samples thus might create noises and ADASYN which focus generating samples that is close to wrongly classified samples thus might accidentally increase chance of overfit (focus on outliers) (imblearn documentation about Over sampling \[10\]).

```python
# import training preprocessing tool from sklearn
from sklearn import preprocessing
from imblearn.over_sampling import SMOTE

# by default, it would only over sample minority class which is the aim in this case
oversample = SMOTE()

# get all features
all_features = train_dataframe.drop("Sepssis", axis=1)
# get labels (Sepssis column)
label = train_dataframe["Sepssis"]

# get list of columns for traversing later on
columns = all_features.columns

# save non sampled data for comparision later on
non_sampled = all_features
non_sampled_label = label

# over sample data set
all_features, label = oversample.fit_resample(all_features, label)

# get univariate feature
uni_feature = all_features[["PL"]]
# feature for tree does not require normalization
tree_all_features = np.array(all_features)

# preprocessing each training set (normalize each feature)
non_sampled = sk.preprocessing.normalize(non_sampled, axis=0)
uni_feature = sk.preprocessing.normalize(uni_feature, axis=0)
all_features = sk.preprocessing.normalize(all_features, axis=0)

print("Size of univariate training data: ", uni_feature.shape)
print("Size of multivariate training data: ", all_features.shape)
print("Size of tree multivariate training data: ", tree_all_features.shape)
```

---
### Model evaluation method

As mentioned in **Model evaluation methods** section aboved, the metrics used for evaluating model would be **Precision, Recall, Accuracy, F1, ROC AUC, confusion matrix**

```python
def evaluating_model(actual, prediction, tree=False, show_result=True):
    """
        :param actual: the actual y value/ label
        :param prediction: the prediction that would be compared against actual value
        :param tree: determine whether to print loss function or not
        :param show_result: whether to print out classification report or not
        :return: dictionary containing all metrics (f1, recall, precision, accuracy) for each class
    """
    classification_stats = sk.metrics.classification_report(actual, prediction, zero_division=0, output_dict=True)
    
    if show_result:
         # print out precision, recall and F1 value
        print(sk.metrics.classification_report(actual, prediction, zero_division=0))

        # print confusion matrix
        print("\nConfusion matrix: ")
        print(sk.metrics.confusion_matrix(actual, prediction))
        
        # print roc_auc score
        print("\nROC AUC score: ")
        print(sk.metrics.roc_auc_score(actual, prediction))

        # print log loss
        if not tree:
            print("\n Log loss")
            print(sk.metrics.log_loss(actual, prediction))   
            
    # add roc_auc to classification stats
    classification_stats["roc_auc"] = sk.metrics.roc_auc_score(actual, prediction)
    
    return classification_stats
```

---

### K-fold validation 
As mentioned previously, K-fold validation is used and in this particular problem. Since the data size is quite small (782 samples) hence the number of fold is kept low and in this case, **8** is the number of fold being applied

```python
# define k-fold that would split data to 8 sets
no_splits = 8
k_fold = sk.model_selection.KFold(n_splits=no_splits, shuffle=True)
```

---

### Model training util    
Since we would need to test out multiple models, there would be multiple duplicated code (for k-fold validation, fitting, printing out result), this function is used for avoid it.

```python
def training_util(model, training_data, label, grid_search=None, tree=False, 
                  feature_imp=True, show_result=True, show_avg=True):
    """
        :param model: model object that would be trained
        :param training_data: input training data that would be feeded to the model
        :param label: input label that would be feeded to the model
        :param grid_search: GridSearchCV object to find the optimal parameters
        :param tree: determine whether inputted model is a decision tree or not (if not, Loss function is printed)
        :param feature_imp: whether to print coefficients/feature importance or not
        :param show_result: whether to show classification report ((precision, recall, F1, confusion matrix)) 
                of predictions or not
        :param show_avg: whether to show average values of metrics (precision, recall, F1, confusion matrix) or not
    """
#     array storing metrics each fold to calculate average
    p_f1 = np.array([])
    n_f1 = np.array([])
    p_recall = np.array([])
    n_recall = np.array([])
    p_precision = np.array([])
    n_precision = np.array([])
    accuracy = np.array([])
    roc_auc = np.array([])
    
#     count which fold being assessed
    fold_counter = 1
    
    for train_index, validate_index in k_fold.split(training_data, label):
        # further splitting data to train and validate data for cross validation
        train_x = training_data[train_index]
        validate_x = training_data[validate_index]
        train_y = label[train_index]
        validate_y = label[validate_index]
        
        # print fold counter
        print(f">>FOLD {fold_counter}/{no_splits}")
        fold_counter+=1

        # feed training data to model
        model.fit(train_x, train_y)

        # make prediction using model
        prediction = model.predict(validate_x)
        
        if show_result:
            # header if result needed to be printed out
            print("\nResult of model")
        result = evaluating_model(validate_y, prediction, tree=tree, show_result=show_result)

        # save f1, recall, precision, accuracy result 
        p_f1=np.append(p_f1, result["1"]["f1-score"])
        n_f1=np.append(n_f1, result["0"]["f1-score"])
        p_recall=np.append(p_recall,result["1"]["recall"])
        n_recall=np.append(n_recall,result["0"]["recall"])
        p_precision=np.append(p_precision,result["1"]["precision"])
        n_precision=np.append(n_precision,result["0"]["precision"])
        accuracy=np.append(accuracy,result["accuracy"])
        roc_auc=np.append(roc_auc,result["roc_auc"])
        
        # check if grid search object is passed (tuning is applied or not)
        if grid_search is not None:
            # fine tuning parameters to get the best ones
            grid_result = grid_search.fit(train_x, train_y)
            print("best params: ", grid_search.best_params_)
            
            if show_result:
                # predicted with tuned model
                optimized_pred = grid_result.predict(validate_x)

                # print result of tuned model
                print("\nResult of optimized model")
                evaluating_model(validate_y, optimized_pred, tree=tree, show_result=show_result)
        
        # print important features if required
        if feature_imp:
            if tree:
                for index, value in enumerate(model.feature_importances_):
                    print(f"Feature {index + 1} have importance of: {value} ")
            else:

                for index, value in enumerate(model.coef_[0]):
                    print(f"Feature {index + 1} have coefficient: {value} ")

        # separator
        print("----"*15)
    print("Completed validation")
    
#     Compute average 
    if show_avg:
        print(f"Average F1-score for Positive: {np.average(p_f1)}")
        print(f"Average F1-score for Negative: {np.average(n_f1)}")
        print(f"Average Recall for Positive: {np.average(p_recall)}")
        print(f"Average Recall for Negative: {np.average(n_recall)}")
        print(f"Average Precision for Positive: {np.average(p_precision)}")
        print(f"Average Precision for Negative: {np.average(n_precision)}")
        print(f"Average Accuracy: {np.average(accuracy)}")
        print(f"Average ROC AUC: {np.average(roc_auc)}")
```

### b) Polynomial Univariate Logistic Regression

As Linear model did not yield a great result, polynomial univariate model would be tested to see whether increase the complexity of the model would yield a better result.  

To select a suitable degree, a graph comparing the performance of model (training and validation) as the polynomial degree would be used for reference. The degree being selected is the one at which the difference between training and validation is low and both value must be high.

```python
train_f1 = list()
val_f1 = list()
poly_degree = np.arange(1, 11, 1)

for i in poly_degree:
    poly_feature = PolynomialFeatures(i)
    uni_poly_feature = poly_feature.fit_transform(uni_feature)

    uni_poly_model = sk.linear_model.LogisticRegression(C=100, penalty='l2', solver="saga", class_weight="balanced", max_iter=200)
    
    x_train, x_test, y_train, y_test = train_test_split(uni_feature, label, test_size=0.2, shuffle=True)
    uni_poly_model.fit(x_train, y_train)
    
    train_pred = uni_poly_model.predict(x_train)
    train_f1.append(f1_score(train_pred, y_train))
    
    test_pred = uni_poly_model.predict(x_test)
    val_f1.append(f1_score(test_pred, y_test))
    
#  plot F1 score for train and validation data set as max_depth increases
plt.plot([deg for deg in poly_degree], [tp for tp in train_f1])
plt.plot([deg for deg in poly_degree], [vp for vp in val_f1])

plt.ylabel("F1 score")
plt.xlabel("Poly degree")
plt.legend(["Training", "Validation"])
plt.show()
```

Looking at the chart, it can be seen that the performance of model does well at degree 3 so it would be the degree applied for comparison

## IV) Tuning Model

From the previous section, we have concluded that **Random Forest** and **L1**, **Multivariate**, **Logistic Regression** yield the best result amongst models being tested, hence these are the model we would use and try to improve by tuning

---
### Tuning method

The method used for tuning in this case is Grid Search. As stated from a paper by Gianluca Malato \[11\], grid search is an exhaustive algorithmn that test out every combinations of values of parameters given and return the best performing combination of value.

If doing grid search on all of the parameters at once takes too much computational time (over 3 parameters), multiple grid searches (a subset of parameters each time) will be applied. 

When the suggested value from Grid Search varies too much each fold or the performance does not improve (or drops) compared to the default model (model without tuning), the range of parameter being tested would be adjusted by first plotting the graph showing performance change when the value of that parameter change. 

---

## <font color="#4B92AF"> A) Tune logistics regression  
    
### Tuning range
    
For **Logistic Regression** model, the following parameters are tunable (definition from sklearn documentation  \[12\])
> - **C** - Inverse of regularization strength, meaning that the smaller the value, the more penalized the model is
> - **max_iter** - the maximum number of iterations taken for solvers to converge

```python
"""
    Parameter range for Logistic Regression
"""
# set the range of lambda value (regularization strength)
lambda_range = np.logspace(-5, 1, num=25)

# set range of C value which is 1/lambda (inverse of regularization strength)
C_range = 1/lambda_range 
max_iter = np.arange(100, 1000, 50)
```

### 1) Tuning C value

```python
lr_model = sk.linear_model.LogisticRegression(C=100,penalty="l1", solver="saga", class_weight="balanced", max_iter=200)

param_range_map = {"C": C_range}

grid_search = sk.model_selection.GridSearchCV(lr_model,param_range_map)
training_util(lr_model, all_features, label, grid_search=grid_search, feature_imp=False, show_avg=False)
```

--- 
Looking at the suggested parameters, it seems that most folds the suggested **C** is around 5.62 hence it is the value being used.

---

### 2) Choosing Max Iteration value

```python

lr_model = sk.linear_model.LogisticRegression(C=5.62,penalty="l1", solver="saga", class_weight="balanced", max_iter=200)

param_range_map = {"max_iter": max_iter}

grid_search = sk.model_selection.GridSearchCV(lr_model,param_range_map)
training_util(lr_model, all_features, label, grid_search=grid_search, feature_imp=False, show_avg=False)
```

---
### <font color="green">*Analysis*</font>
- Looking at the result between the tuned model (optimized model) and model without tuning (default model), little to no changes can be detected. Hence, in this case, it can be concluded that the perfomance of model is the best and in the iteration range of **100 - 350**. To make sure that all coefficients converge, **350** is used


--- 

### 3) Evaluate finalized **Logistic Regression** model

```python
lr_model = sk.linear_model.LogisticRegression(max_iter=350, C = 5.62,
                                      class_weight = "balanced", solver="saga", penalty="l1")

training_util(lr_model, all_features, label, feature_imp=False)
```

### <font color="green">*Analysis*</font>

- Even with tuned value, although the metrics are slightly better, Logistic Regression cannot perform as well with untuned **Random Forest** (especially with the values being maximized **recall** for **positive** and **precision** for **negative**) thus, it is most likely that we would go with **Random Forest** as the final model for this problem

---

## <font color="#4B92AF"> B) Tune Random Forest

### Tuning range 

For **Random Forest** the following parameters are tunable (definition from sklearn documentation \[13\])

> - **cpp_alpha** - Complexity parameter used for Minimal Cost-Complexity Pruning.
> - **min_impurity_decrease** - A node will be split if this split induces a decrease of the impurity greater than or equal to this value.
> - **max_depth** - maximum depth of each tree
> - **min_samples_split** - min number of samples needed to split a node (if **int** then it is the minimum number of samples to split, if **float** then ceil(min_samples_split * n_samples) is the minimum number of samples)
> - **min_samples_leaf** - min number of samples required to be at leaf node (if **int** then it is the minimum number of samples required, if **float** then ceil(min_samples_leaf * n_samples) is the minimum number of samples)


to improve the efficiency for tuning, n_estimators is set to 20

```python
"""
Parameter range for random forest
"""
alpha_range = [0, 0.001, 0.01, 0.1]
max_depth = np.arange(1,101,1)

# test out both int value and float values
min_samples_split = [2, 5, 10, 0.01, 0.05, 0.1, 0.2]
min_samples_leaf = [1, 2, 5, 10, 0.01, 0.05, 0.1, 0.2]
min_impurity_decrease = [0, 0.001, 0.005, 0.01, 0.05, 0.1, 0.2]
```

---

### 1) Tune ccp_alpha and min_impurity_decrease

```python
rf_model = RandomForestClassifier(n_estimators=20, ccp_alpha = 0.01, class_weight = "balanced")

param_range_map = {
                    "ccp_alpha": alpha_range,
                    "min_impurity_decrease":min_impurity_decrease
                    }

grid_search = sk.model_selection.GridSearchCV(rf_model,param_range_map)

training_util(rf_model, tree_all_features, label, grid_search=grid_search,tree=True, feature_imp=False, show_result=True, show_avg=False)
```

--- 
### <font color="green">*Analysis*</font>
- The majority of folds shows that the best params combination is 0 - 0.001 for **ccp_alpha** and 0 - 0.001 for **min_impurity_decrease**. 
- However, going by majority voting, the value would be used for **ccp_alpha is 0.001** and **min_impurity is 0**
---

### 2) Tune min_samples_split and min_samples_leaf

```python
rf_model = RandomForestClassifier(n_estimators=20, ccp_alpha = 0.001, min_impurity_decrease=0, class_weight = "balanced")

param_range_map = {
                    "min_samples_split": min_samples_split,
                    "min_samples_leaf": min_samples_leaf
                    }    


grid_search = sk.model_selection.GridSearchCV(rf_model,param_range_map, scoring=["f1", "precision", "recall", "accuracy"], refit="f1")

training_util(rf_model, tree_all_features, label, tree=True, grid_search=grid_search, feature_imp=False, show_result=True, show_avg=False)
```

--- 
### <font color="green">*Analysis*</font>
- Overall, most of the tuned model performed slightly better than the default model in some or all metrics (since it also trying to balance out recall - precision to maximize F1). 
- Amongst the suggested parameters, the most suggested combinations are **min_samples_leaf** at 1 and **min_samples_split** at 5 which are the value to be used

---

### 3) Tune max_depth

```python
rf_model = RandomForestClassifier(n_estimators=20, ccp_alpha = 0.001, min_impurity_decrease=0, min_samples_leaf=1, min_samples_split = 5,  class_weight = "balanced")

param_range_map = {
                    "max_depth":max_depth
                    }


grid_search = sk.model_selection.GridSearchCV(rf_model,param_range_map)

training_util(rf_model, tree_all_features, label, grid_search=grid_search, tree=True, feature_imp=False, show_result=True, show_avg=False)
```

--- 
### <font color="green">*Analysis*</font>
- As seen from the result from each fold, the change in max_depth does not contribute much as there is little to no change between tuned and un-tuned model.
- However, going by majority voting, max_depth value being chosen is **10**
---

### 4) Choose n_estimators

To make sure that there is a balance between computational cost and the model, a suitable value for n_estimators is selected. Unlike other parameters, as there is no basis for a range in value to apply for Grid Search, the traditional technique using a graph would be used to show the correlation between n_estimators and the score.

```python
for i in range(5): #draw 5 graphs for good measure
    # Define n_estimators range
    n_estimators = [20, 50, 100, 200, 300, 400, 500]

    # variable storing scores over different n_estimators
    train_f1 = list()
    val_f1 = list()

    for i in n_estimators:
        # define model
        rf_model = RandomForestClassifier(n_estimators=i, ccp_alpha = 0.001,
                                          class_weight = "balanced", min_samples_leaf=1, min_samples_split =5,
                                          min_impurity_decrease=0, max_depth=10)

        # generate train and validate data
        x_train, x_val, y_train, y_val = sk.model_selection.train_test_split(tree_all_features, label, test_size=0.2)

        # feed data to model
        rf_model.fit(x_train, y_train)

        # predict using the train data and add F1 score to the list 
        train_pred = rf_model.predict(x_train)
        new_train_f1 = f1_score(y_train, train_pred, average="macro")
        train_f1.append(new_train_f1)

        # predict using the validation data and add F1 score to the list
        val_pred = rf_model.predict(x_val)
        new_val_f1 = f1_score(y_val, val_pred, average="macro")
        val_f1.append(new_val_f1)

    plt.plot([n_est for n_est in n_estimators], [tp for tp in train_f1])
    plt.plot([n_est for n_est in n_estimators], [vp for vp in val_f1])

    # plt.xscale("log")
    plt.ylabel("F1 score")
    plt.xlabel("Number of estimators")
    plt.legend(["Training", "Validation"])
    plt.show()
```

---
### <font color="green">*Analysis*</font>

- Although there is no clear trend/correlation in any of the graph, it is shown that F1 for validation tends to go down around 400-500
- From the graphs, 100 is the lowest estimators that have the most consistent performance hence it is the value being used for the model
---

### 5) Final model evaluation
Lastly, the model with all parameters from previous experiments would be accessed via 8 folds

```python
final_rf_model = RandomForestClassifier(n_estimators=100, ccp_alpha = 0.001, class_weight = "balanced", 
                                        min_samples_leaf=1, min_samples_split = 5,
                                        min_impurity_decrease=0, max_depth=10)


training_util(rf_model, tree_all_features, label, tree=True, feature_imp=False)
```
